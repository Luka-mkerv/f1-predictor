```python
races_list = pd.read_sql('SELECT season, round, race_id FROM races', engine)
```

```python
races_list.head()
```

```python
for idx, row in races_list.iterrows():
    print(row['season'], row['round'], row['race_id'])
```

```python
results = []
overtakes_df = pd.DataFrame(columns=['race_id', 'driver', 'overtake_count'])
```

```python
import time

fastf1.Cache.enable_cache('../data/cache')
fastf1.set_log_level('ERROR')

completed_race_ids = set(overtakes_df['race_id'].unique())

for idx, row in races_list.iterrows():
    if row['race_id'] in completed_race_ids:
        continue
    
    try:
        session = fastf1.get_session(row['season'], row['round'], 'R')
        session.load(messages=False, weather=False)
        
        laps = session.laps
        laps = laps[laps['TrackStatus'] == '1'].copy()
        
        laps['prev_position'] = laps.groupby('Driver')['Position'].transform(lambda x: x.shift(1))
        laps['is_overtake'] = (laps['Position'] < laps['prev_position']) & (laps['PitInTime'].isna()) & (laps['PitOutTime'].isna())
        
        driver_overtakes = laps.groupby('Driver')['is_overtake'].sum()
        
        for driver, count in driver_overtakes.items():
            results.append({'race_id': row['race_id'], 'driver': driver, 'overtake_count': count})
        
        completed_race_ids.add(row['race_id'])
        print(f"Done {row['season']} R{row['round']}")
        time.sleep(15)
        
    except Exception as e:
        print(f"Failed {row['season']} R{row['round']}: {e}")
        time.sleep(60)

overtakes_df = pd.DataFrame(results)
print(overtakes_df['race_id'].nunique(), "races completed")
```

```python
overtakes_df = overtakes_df.sort_values(['driver', 'race_id']).reset_index(drop=True)
overtakes_df['driver_avg_overtakes'] = overtakes_df.groupby('driver')['overtake_count'].transform(lambda x: x.shift(1).expanding().mean())
```

```python
grid_df = pd.read_sql("SELECT race_id, driver_id, grid_position FROM race_results", engine)
overtakes_df = overtakes_df.merge(grid_df, on=['race_id', 'driver_id'], how='left')
print(overtakes_df.shape)
overtakes_df.head()
```

```python
overtakes_df[['overtake_count', 'driver_avg_overtakes', 'grid_position']].isnull().sum()
```

```python
model_df = overtakes_df.dropna(subset=['driver_avg_overtakes']).copy()
print(model_df.shape)
```

```python
seasons_df = pd.read_sql("SELECT race_id, season FROM races", engine)
model_df = model_df.merge(seasons_df, on='race_id', how='left')

train_df = model_df[model_df['season'] < 2025].copy()
test_df = model_df[model_df['season'] == 2025].copy()

print(train_df.shape, test_df.shape)
```

```python
import xgboost as xgb

features = ['grid_position', 'driver_avg_overtakes']

X_train = train_df[features]
X_test = test_df[features]
y_train = train_df['overtake_count']
y_test = test_df['overtake_count']

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'objective': 'count:poisson',
    'eval_metric': 'poisson-nloglik',
    'eta': 0.05,
    'max_depth': 4,
    'min_child_weight': 5,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
}

bst = xgb.train(
    params,
    dtrain,
    num_boost_round=500,
    evals=[(dtrain, 'train'), (dtest, 'test')],
    early_stopping_rounds=40,
    verbose_eval=20
)

print("Best iteration:", bst.best_iteration)
print("Best test score:", bst.best_score)
```

```python
import numpy as np

preds = bst.predict(dtest)
test_df['predicted_overtakes'] = preds

mae = np.abs(test_df['overtake_count'] - test_df['predicted_overtakes']).mean()
print(f"Mean absolute error: {mae:.2f} overtakes")

# also look at actual vs predicted distribution
print(f"\nActual overtakes - mean: {test_df['overtake_count'].mean():.2f}, std: {test_df['overtake_count'].std():.2f}")
print(f"Predicted overtakes - mean: {test_df['predicted_overtakes'].mean():.2f}, std: {test_df['predicted_overtakes'].std():.2f}")
```

```python
baseline_pred = y_train.mean()
baseline_mae = np.abs(test_df['overtake_count'] - baseline_pred).mean()
print(f"Baseline MAE (predict mean always): {baseline_mae:.2f} overtakes")
```

```python
# get constructor avg from database
constructor_df = pd.read_sql("SELECT race_id, driver_id, constructor_id FROM race_results", engine)
constructor_stats = pd.read_sql("""
    SELECT rr.race_id, rr.driver_id, 
           AVG(rr.finish_position) OVER (
               PARTITION BY rr.constructor_id 
               ORDER BY r.date
               ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING
           ) as constructor_career_avg
    FROM race_results rr
    JOIN races r ON rr.race_id = r.race_id
""", engine)

model_df2 = model_df.merge(constructor_stats, on=['race_id', 'driver_id'], how='left')

# train/test split
train_df2 = model_df2[model_df2['season'] < 2025].copy()
test_df2 = model_df2[model_df2['season'] == 2025].copy()

features2 = ['grid_position', 'driver_avg_overtakes', 'constructor_career_avg']

X_train2 = train_df2[features2].fillna(train_df2[features2].mean())
X_test2 = test_df2[features2].fillna(test_df2[features2].mean())

dtrain2 = xgb.DMatrix(X_train2, label=train_df2['overtake_count'])
dtest2 = xgb.DMatrix(X_test2, label=test_df2['overtake_count'])

bst2 = xgb.train(
    params,
    dtrain2,
    num_boost_round=500,
    evals=[(dtrain2, 'train'), (dtest2, 'test')],
    early_stopping_rounds=40,
    verbose_eval=20
)

preds2 = bst2.predict(dtest2)
test_df2['predicted_overtakes'] = preds2
test_df2['predicted_rounded'] = test_df2['predicted_overtakes'].round()

exact2 = (test_df2['predicted_rounded'] == test_df2['overtake_count']).sum()
within2 = (np.abs(test_df2['predicted_rounded'] - test_df2['overtake_count']) <= 2).sum()
total2 = len(test_df2)

print(f"Exact match: {exact2}/{total2} ({exact2/total2*100:.1f}%)")
print(f"Within 2: {within2}/{total2} ({within2/total2*100:.1f}%)")
```
